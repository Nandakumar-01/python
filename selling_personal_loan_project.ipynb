```python
for col in df.columns:
    q25 = df[col].quantile(.25)
    q75 = df[col].quantile(.75)
    iqr = q75-q25

    min_value = q25-(1.5*iqr)
    max_value = q75+(1.5*iqr)

    df.loc[(df[col]< min_value) & (df[col]> max_value),col] = df[col].quantile(.50)

    print(col,df[col].isnull().sum())
```

```python
x = df.loc[:,df.columns !="Personal Loan"]
y = df.loc[:,df.columns == "Personal Loan"]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.20,random_state=5)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
model_score = model.score(x_test,y_test)
print(model_score)
```

```python
from sklearn import metrics

print(metrics.confusion_matrix(y_test,y_predict))
```

```python
print(metrics.classification_report(y_test,y_predict))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
knn_predict = knn.predict(x_test)
metrics.confusion_matrix(y_test,knn_predict)
```

```python
print(metrics.classification_report(y_test,knn_predict))
```

```python
from sklearn import preprocessing

x_train_scale = preprocessing.scale(x_train)
x_test_scale = preprocessing.scale(x_test)
```

```python
knn.fit(x_train_scale,y_train)
```

```python
predicted = knn.predict(x_test_scale)

metrics.confusion_matrix(y_test,predicted)
```

```python
print(metrics.classification_report(y_test,predicted))
```

```python
model = LogisticRegression()
model.fit(x_train_scale,y_train)

model_score = model.score(x_test_scale,y_test)

y_predict = model.predict(x_test_scale)

print(model_score)
print(metrics.confusion_matrix(y_test,y_predict))
```

```python
print(metrics.classification_report(y_test,y_predict))
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier()
tree_model.fit(x_train_scale,y_train)
```

```python
y_pred = tree_model.predict(x_test_scale)
```

```python
print(metrics.confusion_matrix(y_test,y_pred))
print(metrics.classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100,random_state=20)
rf_model.fit(x_train_scale,y_train)
```

```python
y_pred_rf = rf_model.predict(x_test_scale)

print(metrics.confusion_matrix(y_test,y_pred_rf))
print(metrics.classification_report(y_test,y_pred_rf))
```

```python
from sklearn.ensemble import BaggingClassifier

base = DecisionTreeClassifier(max_depth=10)

bagg = BaggingClassifier(base,n_estimators=20,random_state=10)

bagg.fit(x_train_scale,y_train)

y_pred_bagg = bagg.predict(x_test_scale)
```

```python
print(metrics.confusion_matrix(y_test,y_pred_bagg))
print(metrics.classification_report(y_test,y_pred_bagg))
```

```python
from sklearn.ensemble import AdaBoostClassifier
base_model = DecisionTreeClassifier(max_depth=5)
ada = AdaBoostClassifier(base_model,n_estimators=50,random_state=20)
ada.fit(x_train_scale,y_train)
y_pred_ada = ada.predict(x_test_scale)
```

```python
print(metrics.confusion_matrix(y_test,y_pred_ada))
print(metrics.classification_report(y_test,y_pred_ada))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

grade = GradientBoostingClassifier(n_estimators=100,random_state=10)
grade.fit(x_train_scale,y_train)

y_pred_grade = grade.predict(x_test_scale)
```

```python
print(metrics.confusion_matrix(y_test,y_pred_grade))
print(metrics.classification_report(y_test,y_pred_grade))
```
